```python
from bs4 import BeautifulSoup
```

```python
url="https://www.worldometers.info/coronavirus/"
```

```python
import requests
```

```python
contents=requests.get(url).content
```

```python
soup=BeautifulSoup(contents,'html.parser')
```

```python
covid_data=soup.find('tbody').find_all('tr')
```

```python
covid_data[8]
```

```python
whole_data=[]
for item in range(8,len(covid_data)):
    data=[]
    dat=covid_data[item].find_all('td')
    for i in dat:
        text=(i.text)
        data.append(text)
    whole_data.append(data)
```

```python
whole_data
```

```python
cleaned_data=list(map(lambda x: x[1:10] + [x[12],x[14]],whole_data))
```

```python
column_names=["Country",
"Total cases",
"New cases",
"total deaths",
"new deaths",
"total recovered",
"new recovered",
"active cases",
"serious Cases",
"total test",
"population",]
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(cleaned_data,columns = column_names)
```

```python
df.to_csv('coviddata.csv',index=False)
```
